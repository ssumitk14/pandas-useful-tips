```python
df1 = pd.DataFrame({'A': ['x1','x2','x3', 'x4'], 'B':[['v1','v2'],['v3','v4'],['v5','v6'],['v7','v8']], 'C':[['c1','c2'],['c3','c4'],['c5','c6'],['c7','c8']],'D':[['d1','d2'],['d3','d4'],['d5','d6'],['d7','d8']], 'E':[['e1','e2'],['e3','e4'],['e5','e6'],['e7','e8']]})
df1
```

```python
%%time
df1.apply(pd.Series.explode)
```

```python
#Specify columns that you want to modify

df1.apply(lambda x: x.explode() if x.name in ['B', 'C', 'D', 'E'] else x)
```

```python
df1.apply(lambda x: x.explode(ignore_index=True) if x.name in ['B', 'C', 'D', 'E'] else x)
```
